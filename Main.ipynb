```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
def compute_column_statistics(df, column_name):
    df[column_name] = df[column_name].astype(str)
    value_counts = df[column_name].value_counts()
    total_examples = len(df)
    percentages = (value_counts / total_examples) * 100
    return pd.DataFrame(
        {
            'value': value_counts.index,
            'count': value_counts.values,
            'percentage': percentages.values
        }
    )

def plot_column_values(result_df, column_name):
    plt.figure(figsize=(10, 6))
    plt.bar(result_df['value'], result_df['percentage'])
    plt.xlabel(column_name)
    plt.ylabel('Percentage')
    plt.title(f'{column_name} percentages')
    plt.xticks(rotation=45)
    plt.show()
```

## MOL

```python
mol_df: pd.DataFrame = pd.read_csv("mol.csv")
mol_df.columns
```

```python
for column in ['explicit-or-implicit', 'pt-contextual-label', 'pt-hate-label']:
    column_statistics = compute_column_statistics(mol_df, column)
    print(column_statistics)
    plot_column_values(column_statistics, column)
```
